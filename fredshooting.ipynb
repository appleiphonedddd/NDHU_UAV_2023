```python
#按Enter鍵後4部WEBCAM拍照1張照片
import cv2
import os
from datetime import datetime

# 初始化四個Webcam的視訊擷取對象
cap1 = cv2.VideoCapture(0)
cap2 = cv2.VideoCapture(1)
cap3 = cv2.VideoCapture(2)
cap4 = cv2.VideoCapture(3)

# 檢查是否成功開啟視訊擷取對象
if not cap1.isOpened() or not cap2.isOpened() or not cap3.isOpened() or not cap4.isOpened():
    print("無法開啟一個或多個Webcam。")
    exit()

# 建立儲存照片的目錄
output_dir = 'D:/UAV'
os.makedirs(output_dir, exist_ok=True)

# 初始化照片計數器
photo_count = 1

# 列印程式已啟動訊息
print("程式已啟動，按下 'Enter' 鍵開始拍照，再次按 'Enter' 鍵暫停拍照。")

recording = False

while True:
    # 等待用户按下 Enter 键
    if input() == '':
        recording = not recording

    if recording:
        # 從四個Webcam讀取一幀視頻
        ret1, frame1 = cap1.read()
        ret2, frame2 = cap2.read()
        ret3, frame3 = cap3.read()
        ret4, frame4 = cap4.read()

        # 如果成功讀取視頻，保存照片
        if ret1:
            # 建立照片檔案名，包括日期、時間和毫秒
            now = datetime.now()
            timestamp = now.strftime("%Y%m%d_%H%M%S_%f")
            filename1 = f'{output_dir}/UAV1/Webcam1_{timestamp}.jpg'

            # 保存照片
            cv2.imwrite(filename1, frame1)

            # 列印已儲存的照片檔名
            print(f"儲存照片: {filename1}")

        if ret2:
            now = datetime.now()
            timestamp = now.strftime("%Y%m%d_%H%M%S_%f")
            filename2 = f'{output_dir}/UAV2/Webcam2_{timestamp}.jpg'
            cv2.imwrite(filename2, frame2)
            print(f"儲存照片: {filename2}")

        if ret3:
            now = datetime.now()
            timestamp = now.strftime("%Y%m%d_%H%M%S_%f")
            filename3 = f'{output_dir}/UAV3/Webcam3_{timestamp}.jpg'
            cv2.imwrite(filename3, frame3)
            print(f"儲存照片: {filename3}")

        if ret4:
            now = datetime.now()
            timestamp = now.strftime("%Y%m%d_%H%M%S_%f")
            filename4 = f'{output_dir}/UAV4/Webcam4_{timestamp}.jpg'
            cv2.imwrite(filename4, frame4)
            print(f"儲存照片: {filename4}")

# 釋放資源並關閉視窗
cap1.release()
cap2.release()
cap3.release()
cap4.release()
cv2.destroyAllWindows()
print("程式已退出。")
```

```python
#按Enter鍵後4部WEBCAM拍照10張照片，時間一樣、序號不一樣
import cv2
import os
from datetime import datetime

# 初始化四個Webcam的視訊擷取對象
cap1 = cv2.VideoCapture(0)
cap2 = cv2.VideoCapture(1)
cap3 = cv2.VideoCapture(2)
cap4 = cv2.VideoCapture(3)

# 檢查是否成功開啟視訊擷取對象
if not cap1.isOpened() or not cap2.isOpened() or not cap3.isOpened() or not cap4.isOpened():
    print("無法開啟一個或多個Webcam。")
    exit()

# 建立儲存照片的目錄
output_dir = 'D:/UAV'
os.makedirs(output_dir, exist_ok=True)

# 列印程式已啟動訊息
print("程式已啟動，按下 'Enter' 鍵開始拍照10張照片,再次按 'Enter' 鍵停止。按 'Q' 键退出程序。")

recording = False
photo_limit = 10
photos_taken = 0

while True:
    # 等待用户按下 Enter 鍵
    if input() == '':
        if not recording:
            # 獲取當前的日期、時間和毫秒，用於建立文件名
            now = datetime.now()
            timestamp = now.strftime("%Y%m%d_%H%M%S_%f")
            photo_info = f'Webcam_{timestamp}'
            recording = True
            photos_taken = 0

    if recording:
        for _ in range(photo_limit):
            # 從四個Webcam讀取一幀視頻
            ret1, frame1 = cap1.read()
            ret2, frame2 = cap2.read()
            ret3, frame3 = cap3.read()
            ret4, frame4 = cap4.read()

            if ret1:
                filename1 = f'{output_dir}/UAV1/{photo_info}_Cam1_{photos_taken + 1}.jpg'
                cv2.imwrite(filename1, frame1)
                print(f"儲存照片: {filename1}")

            if ret2:
                filename2 = f'{output_dir}/UAV2/{photo_info}_Cam2_{photos_taken + 1}.jpg'
                cv2.imwrite(filename2, frame2)
                print(f"儲存照片: {filename2}")

            if ret3:
                filename3 = f'{output_dir}/UAV3/{photo_info}_Cam3_{photos_taken + 1}.jpg'
                cv2.imwrite(filename3, frame3)
                print(f"儲存照片: {filename3}")

            if ret4:
                filename4 = f'{output_dir}/UAV4/{photo_info}_Cam4_{photos_taken + 1}.jpg'
                cv2.imwrite(filename4, frame4)
                print(f"儲存照片: {filename4}")

            photos_taken += 1
        
        recording = False

    key = cv2.waitKey(1) & 0xFF

    if key == ord('q'):
        print("程式已退出。")
        break

# 释放资源并关闭窗口
cap1.release()
cap2.release()
cap3.release()
cap4.release()
cv2.destroyAllWindows()
```

```python
#自動攝影方式
import cv2
import os
from datetime import datetime

# 初始化四個Webcam的視訊擷取對象
cap1 = cv2.VideoCapture(0)
cap2 = cv2.VideoCapture(1)
cap3 = cv2.VideoCapture(2)
cap4 = cv2.VideoCapture(3)

# 檢查是否成功開啟視訊擷取對象
if not cap1.isOpened() or not cap2.isOpened() or not cap3.isOpened() or not cap4.isOpened():
    print("无法打开一个或多个Webcam")
    exit()

# 設定視訊編碼器
fourcc = cv2.VideoWriter_fourcc(*'XVID')

# 建立四個VideoWriter對象，用於保存四個Webcam的錄製視頻
output_dir = 'D:/UAV/'
os.makedirs(output_dir, exist_ok=True)

code = 'UAV001'  # # 代號
out1 = cv2.VideoWriter(f'{output_dir}{code}_{datetime.now().strftime("%Y%m%d_%H%M%S")}_Webcam1.avi', fourcc, 20.0, (640, 480))
out2 = cv2.VideoWriter(f'{output_dir}{code}_{datetime.now().strftime("%Y%m%d_%H%M%S")}_Webcam2.avi', fourcc, 20.0, (640, 480))
out3 = cv2.VideoWriter(f'{output_dir}{code}_{datetime.now().strftime("%Y%m%d_%H%M%S")}_Webcam3.avi', fourcc, 20.0, (640, 480))
out4 = cv2.VideoWriter(f'{output_dir}{code}_{datetime.now().strftime("%Y%m%d_%H%M%S")}_Webcam4.avi', fourcc, 20.0, (640, 480))

recording = False

while True:
    # 從四個Webcam讀取一幀視頻
    ret1, frame1 = cap1.read()
    ret2, frame2 = cap2.read()
    ret3, frame3 = cap3.read()
    ret4, frame4 = cap4.read()

    # 如果成功讀取視頻，則寫入錄製視頻
    if recording:
        if ret1:
            out1.write(frame1)
        if ret2:
            out2.write(frame2)
        if ret3:
            out3.write(frame3)
        if ret4:
            out4.write(frame4)

    # 顯示來自四個Webcam的視頻
    cv2.imshow('Webcam 1', frame1)
    cv2.imshow('Webcam 2', frame2)
    cv2.imshow('Webcam 3', frame3)
    cv2.imshow('Webcam 4', frame4)

    # 按下'q'鍵退出錄製
    key = cv2.waitKey(1)
    if key == ord('q'):
        break
    # 按下's'鍵開始/停止錄製
    elif key == ord('s'):
        recording = not recording

# 釋放資源
cap1.release()
cap2.release()
cap3.release()
cap4.release()
out1.release()
out2.release()
out3.release()
out4.release()
cv2.destroyAllWindows()
```
